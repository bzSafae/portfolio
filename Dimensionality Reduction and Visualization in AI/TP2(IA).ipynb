```python
# Sparation des features (gnes) et de la variable cible (tissue_status)
X = data.drop(columns=['Athlets','Rank','Points','Competition'])  # Features
y = data[['Rank','Points','Competition']]  # Variable cible


X.head()
```

```python
#Standardisation

import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler


# Standardisation avec MinMaxScaler (normalisation entre 0 et 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Convertir les donnes standardises en DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Ajouter la colonne 'tissue_status'  nouveau pour la variable cible
X_scaled_df[['Rank','Points','Competition']] = y

# Afficher un aperu des donnes standardises
print(X_scaled_df.head())


print('\n********** Standardisation*********\n')
```

```python
from sklearn.decomposition import PCA

# Ralisation de l'ACP
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_scaled)

# Affichage de l'explication de la variance par composante
print(pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt

explained_varience=pca.explained_variance_ratio_

for i,variance in enumerate(explained_varience):
  print(f'variance expliqu par la composante {i+1}:{variance:.4f}')

# Diagramme en btons des valeurs propres
plt.bar(range(1, 11), pca.explained_variance_ratio_)
plt.xlabel('Composantes principales')
plt.ylabel('Variance explique')
plt.title('Diagramme en btons des valeurs propres')
plt.show()
```

```python
variance_cumulee = pca.explained_variance_ratio_.cumsum()
print(variance_cumulee)
```

```python
# Importer les bibliothques ncessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Appliquer l'ACP avec les 2 premires composantes principales
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Extraire les charges (les coefficients des variables sur les composantes principales)
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

# Tracer le cercle de corrlation
def plot_correlation_circle(loadings, pca, labels):
    plt.figure(figsize=(8, 8))
    plt.quiver(np.zeros(loadings.shape[0]), np.zeros(loadings.shape[0]),
               loadings[:, 0], loadings[:, 1],
               angles='xy', scale_units='xy', scale=1)
    for i, label in enumerate(labels):
        plt.text(loadings[i, 0], loadings[i, 1], label, color='black', ha='center', va='center')

    # Cercles de corrlation
    circle = plt.Circle((0, 0), 1, color='blue', fill=False, linestyle='--')
    plt.gca().add_artist(circle)

    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    plt.axhline(0, color='grey', lw=1)
    plt.axvline(0, color='grey', lw=1)
    plt.xlabel(f"Composante 1 ({pca.explained_variance_ratio_[0]*100:.1f}% de variance explique)")
    plt.ylabel(f"Composante 2 ({pca.explained_variance_ratio_[1]*100:.1f}% de variance explique)")
    plt.title('Cercle de corrlation des variables')
    plt.grid()
    plt.show()

# Reprsentation des individus
def plot_individuals(X_pca, y):
    plt.figure(figsize=(8, 6))
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y['Rank'], cmap='viridis')
    plt.colorbar(label='Classement')
    plt.title('Projection des athltes sur les deux premires composantes principales')
    plt.xlabel(f'Composante 1 ({pca.explained_variance_ratio_[0]*100:.1f}% de variance explique)')
    plt.ylabel(f'Composante 2 ({pca.explained_variance_ratio_[1]*100:.1f}% de variance explique)')
    plt.grid()
    plt.show()

# Tracer le cercle de corrlation
variables = X.columns
plot_correlation_circle(loadings, pca, variables)

# Tracer la projection des individus (athltes)
plot_individuals(X_pca, y)
```

```python
# Importer les bibliothques ncessaires
import matplotlib.pyplot as plt
import seaborn as sns

# tape 1 : Exploration des donnes
# Distribution des points et des rangs
plt.figure(figsize=(10, 6))
sns.histplot(data['Points'], bins=20, kde=True, color='blue')
plt.title('Distribution des Points')
plt.xlabel('Points')
plt.ylabel('Frquence')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data['Rank'], bins=10, kde=True, color='green')
plt.title('Distribution des Classements (Rank)')
plt.xlabel('Rank')
plt.ylabel('Frquence')
plt.grid(True)
plt.show()

# tape 2 : Comparaison entre les comptitions
# Scatterplot des athltes en fonction des points et du classement (Rank), color par la comptition
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Points', y='Rank', hue='Competition', data=data, palette='Set1')

# Ajouter les noms des athltes  chaque point
for i in range(data.shape[0]):
    plt.text(data['Points'][i], data['Rank'][i], data['Athlets'][i],
             fontsize=9, verticalalignment='bottom')

plt.title('Classement des athltes en fonction des points par comptition')
plt.xlabel('Points')
plt.ylabel('Classement (Rank)')
plt.gca().invert_yaxis()  # Inverser l'axe des y pour que le classement 1 soit en haut
plt.grid(True)
plt.show()

# Boxplot pour visualiser la distribution des points par comptition
plt.figure(figsize=(10, 6))
sns.boxplot(x='Competition', y='Points', data=data, palette='coolwarm')
plt.title('Distribution des points par comptition')
plt.xlabel('Comptition')
plt.ylabel('Points')
plt.grid(True)
plt.show()

# tape 3 : Analyse des meilleurs profils
# Extraire les athltes ayant les meilleurs rangs (1er  3e rang)
best_athletes = data[data['Rank'] <= 3]

plt.figure(figsize=(10, 6))
sns.barplot(x='Competition', y='Points', hue='Rank', data=best_athletes, palette='viridis')
plt.title('Meilleurs athltes (1er  3e rang) par comptition')
plt.xlabel('Comptition')
plt.ylabel('Points')
plt.legend(title='Classement')
plt.grid(True)
plt.show()

# tape 4 : Analyse des relations entre les variables
# Heatmap pour observer la corrlation entre Rank, Points et les performances dans les diffrentes preuves
performance_columns = ['100m', 'Long.jump', 'Shot.put', 'High.jump', '400m', '110m.hurdle',
                       'Discus', 'Pole.vault', 'Javeline', '1500m']

plt.figure(figsize=(12, 8))
corr_matrix = data[['Rank', 'Points'] + performance_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Corrlation entre Rank, Points et les performances')
plt.show()

# tape 5 : Visualisation des rangs par preuve
# Scatterplot des points par preuve avec le rang et les noms des athltes
for event in performance_columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=event, y='Rank', data=data, hue='Competition', palette='Spectral')

    # Ajouter les noms des athltes  chaque point
    for i in range(data.shape[0]):
        plt.text(data[event][i], data['Rank'][i], data['Athlets'][i],
                 fontsize=9, verticalalignment='bottom')

    plt.title(f'Classement en fonction de la performance dans {event}')
    plt.xlabel(event)
    plt.ylabel('Classement (Rank)')
    plt.gca().invert_yaxis()
    plt.grid(True)
    plt.show()
```
